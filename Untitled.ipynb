```python
class Students(tk.Toplevel):  
    def __init__(self, root,username,spedata):  
        super().__init__(root)
        self.geometry("1000x600+200+50") 
        self.configure(bg='lightblue') 
        self.title("Class schedules")  
        self.username=username
        self.spedata=spedata
        self.dataname=self.username+'.json'
        with open(self.dataname, 'r') as file:  
            self.data = json.load(file)  
  
        # Creating the Left Column  
        self.option_frame = tk.Frame(self)  
        self.option_frame.pack(side="left", fill="y")  
  
        # Create a button and bind a callback function  
        self.button1 = tk.Button(self.option_frame, text="Current\nWeek", command=lambda: self.switch_content("option_1"))  
        self.button1.pack(fill="x")  
        self.button2 = tk.Button(self.option_frame, text="Semester", command=lambda: self.switch_content("option_2"))  
        self.button2.pack(fill="x")  
  
        # Creating the main page display area  
        self.content_frame = tk.Frame(self)  
        self.content_frame.pack(side="left", fill="both", expand=True)  
  
        # Initialising the display  
        self.switch_content("option_1")  
          
        self.is_closed = False  # 标志位，用于指示窗口是否已关闭  
        self.protocol("WM_DELETE_WINDOW", self.on_closing) 
        
    def on_closing(self):  
        self.is_closed = True  # 标记窗口为已关闭  
        root.destroy()
        
    def switch_content(self, option):  
        # Clear current display  
        for widget in self.content_frame.winfo_children():  
            widget.destroy()  
  
        # 根据选项添加新的内容  
        if option == "option_1":  
            content_for_option1 = tk.Frame(self.content_frame)  
            content_for_option1.pack(fill="both", expand=True)
            content_for_option1.configure(bg='lightblue')
    
            y={}
            for i in range (6):
                label= tk.Label(content_for_option1, text=weekday[i],width=20, height=1)
                label.place(x=i * 150 + 50, y=30)

            for i in range (5):
                label= tk.Label(content_for_option1, text=time[i],width=5, height=5)
                label.place(x=5, y=i*100+60)
            for course in self.data:
                scr = "Name:{}\nCode:{}\nTeacher:{}\nSession:{}\nWeek:{}".format(
                    course["name"],course["code"],course["teacher"], course["session"], course["week"])
                week = course["week"]
                if "odd" in week:
                    week = week.replace("(odd)", "")
                    week = week.split("-")
        # 开始周/结束周
                    start_week, end_week = eval(week[0]), eval(week[-1])
                    if weeks % 2 == 1: # 判断是否是单周
                        if start_week <= weeks <= end_week:  # 判断该课程是否是当前周的课程
                            if start_week <= weeks <= end_week:  # 判断该课程是否是当前周的课程
                                x = weekday.index(course["day"])
                                y[course["day"]] = int(course['session'])-1
                                label = tk.Label(content_for_option1, text=scr,width=20, height=5)  # 创建一个文本控件
                                label.place(x=x * 150 + 50, y=y[course["day"]] * 100 + 60)  # 在屏幕上放置文本控件  # 在控件上放置文本
            
                elif "even" in week:
                    week = week.replace("(even)", "")
                    week = week.split("-")
        # 开始周/结束周
                    start_week, end_week = eval(week[0]), eval(week[-1])
                    if weeks % 2 == 0: # 判断是否是单周
                        if start_week <= weeks <= end_week:  # 判断该课程是否是当前周的课程
                            if start_week <= weeks <= end_week:  # 判断该课程是否是当前周的课程
                                x = weekday.index(course["day"])
                                y[course["day"]] = int(course['session'])-1
                                label = tk.Label(content_for_option1, text=scr,width=20, height=5)  # 创建一个文本控件
                                label.place(x=x * 150 + 50, y=y[course["day"]] * 100 + 60)  # 在屏幕上放置文本控件  # 在控件上放置文本
                else:
                    week = week.split("-")
        # 开始周/结束周
                    start_week, end_week = eval(week[0]), eval(week[-1])
                    if start_week <= weeks <= end_week:  # 判断该课程是否是当前周的课程
                        if start_week <= weeks <= end_week:  # 判断该课程是否是当前周的课程
                            x = weekday.index(course["day"])
                            y[course["day"]] = int(course['session'])-1
                            label = tk.Label(content_for_option1, text=scr,width=20, height=5)  # 创建一个文本控件
                            label.place(x=x * 150 + 50, y=y[course["day"]] * 100 + 60)  # 在屏幕上放置文本控件  # 在控件上放置文本
                    
            #print(scr)

            # 确认文本的位置
                
            # 在选项一的内容容器中添加组件  
            weeklabel="Week:{},{}".format(weeks,weekslabel)
            label=tk.Label(content_for_option1, text=weeklabel,font=("Arial", 12,"bold"),fg="deepskyblue")
            label.place(x=400,y=0) 
      
        elif option == "option_2":  
            content_for_option2 = tk.Frame(self.content_frame)  
            content_for_option2.configure(bg='lightblue') 
            content_for_option2.pack(fill="both", expand=True)  
  
            # 在选项二的内容容器中添加组件  
            y={}
            for i in range (6):
                label= tk.Label(content_for_option2, text=weekday[i],width=20, height=1)
                label.place(x=i * 150 + 50, y=30)
    
            for i in range (5):
                label= tk.Label(content_for_option2, text=time[i],width=5, height=5)
                label.place(x=5, y=i*100+60)
            for course in self.data:
                scr = "Name:{}\nCode:{}\nTeacher:{}\nSession:{}\nWeek:{}".format(
                    course["name"],course["code"],course["teacher"], course["session"], course["week"])
                #print(scr)

            # 确认文本的位置
                x = weekday.index(course["day"])
                y[course["day"]] = int(course['session'])-1

                label = tk.Label(content_for_option2, text=scr,width=20, height=5)  # 创建一个文本控件
                label.place(x=x * 150 + 50, y=y[course["day"]] * 100 + 60)  # 在屏幕上放置文本控件
                
            label=tk.Label(content_for_option2, text='Class schedules for the whole semester',font=("Arial", 12,"bold"),fg="deepskyblue")
            label.place(x=300,y=0) 
            button1 = tk.Button(content_for_option2, text="Add Course", command=lambda: self.course_operations("add"))
            button1.place(x=240,y=560)
            button2 = tk.Button(content_for_option2, text="Delete Course", command=lambda: self.course_operations("delete"))
            button2.place(x=540,y=560)

            
    def course_operations(self, option): 
        if option == "add": 
            self.add_option = tk.Toplevel(root)
            self.add_option.geometry("700x250+500+250")
            self.add_option.title("Add...")
            
            tree = ttk.Treeview(self.add_option, columns=('name', 'code', 'teacher', 'session', 'week','day'), height=7,show="headings")
            tree.heading('name', text="name", anchor=W)
            tree.column('name', width=150,anchor=W)
            tree.heading('code', text="code", anchor=W)
            tree.column('code', width=100, anchor=W)
            tree.heading('teacher', text="teacher", anchor=W)
            tree.column('teacher', width=100, anchor=W)
            tree.heading('session',text="session",anchor=W)
            tree.column('session', width=100, anchor=W)
            tree.heading('week', text="week", anchor=W)
            tree.column('week', width=100, anchor=W)
            tree.heading('day', text="day", anchor=W)
            tree.column('day', width=100, anchor=W)
            
            for course in self.spedata:
                scr = []
                scr.append(course["name"])
                scr.append(course["code"])
                scr.append(course["teacher"])
                scr.append(course["session"])
                scr.append(course["week"])
                scr.append(course["day"])
                tree.insert("",END,values=scr)
            tree.pack(pady=10) 
            label=tk.Label(self.add_option, text='Enter Code:')
            label.place(x=250,y=180)
            self.text_boxx = tk.Entry(self.add_option, width=10)  
            self.text_boxx.place(x=330,y=180)
            
            button1 = tk.Button(self.add_option, text="Add", command=lambda: self.add_course())  
            button2 = tk.Button(self.add_option, text="Cancel", command=lambda:self.close_window(self.add_option))
            
            button1.place(x=200,y=210)   
            button2.place(x=500,y=210)
            
        if option == "delete": 
            #self.delete_option = delete_option
            self.delete_option = tk.Toplevel(root)
            self.delete_option.geometry("280x90+550+300")
            self.delete_option.title("Delete...")
            
            label=tk.Label(self.delete_option, text='Enter Code:')
            label.place(x=20,y=15)
            self.text_box = tk.Text(self.delete_option, height=1, width=10)  
            self.text_box.place(x=100,y=16) 
            
            button1 = tk.Button(self.delete_option, text="Confirm",command=lambda:self.get_text(self.text_box))
            button1.place(x=50,y=45)
            button2 = tk.Button(self.delete_option, text="Cancel",command=lambda:self.close_window(self.delete_option))
            button2.place(x=150,y=45)
    
    def get_text(self,box):  
        target_code =box.get("1.0", "2.0").strip()  
        self.delete_option.destroy()
        for course in self.data: 
            if course.get('code') == target_code:  
                self.data.remove(course)
                messagebox.showinfo(title="Delete successfully", message="Delete successfully!!")
        with open(self.dataname, 'w') as file:  
            json.dump(self.data, file, ensure_ascii=False, indent=4)
        
        self.switch_content("option_2")

    def add_course(self):
        target_code =self.text_boxx.get().strip()
        flag=0
        for course in self.spedata: 
            if course.get('code') == target_code:  
                self.data.append(course)
                flag=1
                messagebox.showinfo(title="Add successfully", message="Added successfully!!")
        if flag==0:
            messagebox.showerror("Add Failed", "The code is incorrect, please try again.")
            self.text_boxx.delete(0, tk.END)
        with open(self.dataname, 'w') as file:  
            json.dump(self.data, file, ensure_ascii=False, indent=4)
            
        self.switch_content("option_2")
        self.add_option.destroy()
    def close_window(self,window):
        window.destroy()
```

```python
class Login(tk.Toplevel):
    def __init__(self, root,userdata):  
        super().__init__(root) 
        self.geometry("300x200+600+300") 
        self.configure(bg='lightblue') 
        self.title("Login")  
        self.userdata=userdata
        
        label=tk.Label(self, text='Username:')
        label.place(x=30,y=15)
        self.text_box = tk.Entry(self, width=20)  
        self.text_box.place(x=100,y=16) 
        label2=tk.Label(self, text='Password:')
        label2.place(x=30,y=45)
        self.text_box2 = tk.Entry(self, width=20,show='*')  
        self.text_box2.place(x=100,y=46) 
            
        button1 = tk.Button(self, text="Confirm",command=lambda:self.check(self.text_box))
        button1.place(x=60,y=80)
        button2 = tk.Button(self, text="Cancel",command=lambda:self.close_window())
        button2.place(x=170,y=80)
        self.is_closed = False  # 标志位，用于指示窗口是否已关闭  
        self.protocol("WM_DELETE_WINDOW", self.on_closing) 
        
    def on_closing(self):  
        self.is_closed = True  # 标记窗口为已关闭  
        root.destroy()
        # Creating the Left Column  
        
    def check(self,box):  
        name =self.text_box.get().strip() 
        password_=self.text_box2.get().strip()
        flag=0
        for student in self.userdata['students']:
            if student.get('username') == name and student.get('password') == password_ :
                main_window=Students(root,name,spedata)
                flag=1
                self.destroy()
        for teacher in self.userdata['teachers'] :
            if teacher.get('username') == name and teacher.get('password') == password_:
                flag=1
                main_window=Teachers(root,name,spedata)
                self.destroy()
        for administrator in self.userdata['administrators']:
            if administrator.get('username') == name and administrator.get('password') == password_:
                main_window=Admins(root,userdata)
                flag=1
                self.destroy()
        if flag==0:
            messagebox.showerror("Login Failed", "The username or password is incorrect, please try again.")
            self.text_box.delete(0, tk.END)
            self.text_box2.delete(0, tk.END)
            
    
    def close_window(self):
        self.destroy()
        root.destroy()
          
  
```

```python
class Admins(tk.Toplevel):
    def __init__(self, root):  
        super().__init__(root) 
        self.geometry("300x200+600+300") 
        self.configure(bg='lightblue') 
        self.title("Admin")  
        self.userdata=userdata
            
        button1 = tk.Button(self, text="Create a new user ",command=lambda:self.add())
        button1.place(x=60,y=50)
        button2 = tk.Button(self, text="Find password",command=lambda:self.find())
        button2.place(x=60,y=100)
        # Creating the Left Column  
        self.is_closed = False  # 标志位，用于指示窗口是否已关闭  
        self.protocol("WM_DELETE_WINDOW", self.on_closing) 
        
    def on_closing(self):  
        self.is_closed = True  # 标记窗口为已关闭  
        root.destroy()
        
    def add(self):
        self.adduser = tk.Toplevel(root)
        self.adduser.geometry("280x150+610+320")
        self.adduser.title("Add...")
            
        label=tk.Label(self.adduser, text='Username:')
        label.place(x=30,y=15)
        self.text_box1 = tk.Entry(self.adduser, width=20)  
        self.text_box1.place(x=100,y=16) 
        label2=tk.Label(self.adduser, text='Password:')
        label2.place(x=30,y=45)
        self.text_box3 = tk.Entry(self.adduser, width=20)  
        self.text_box3.place(x=100,y=46)  
            
        button1 = tk.Button(self.adduser, text="Confirm",command=lambda:self.get_add())
        button1.place(x=50,y=100)
        button2 = tk.Button(self.adduser, text="Cancel",command=lambda:self.close_window(self.adduser))
        button2.place(x=170,y=100)
    
    def get_add(self):  
        name =self.text_box1.get().strip() 
        password_=self.text_box3.get().strip()
        newuser={'username':name, 'password':password_ }  
        if len(name)==11:
            self.userdata['students'].append(newuser)
            messagebox.showinfo(title="Add successfully", message="Added successfully!!")
            with open('userdata.json', 'w') as file:  
                json.dump(self.userdata, file, ensure_ascii=False, indent=4)
            self.adduser.destory()
        elif len(name)==8:
            self.userdata['teachers'].append(newuser)
            messagebox.showinfo(title="Add successfully", message="Added successfully!!")
            with open('userdata.json', 'w') as file:  
                json.dump(self.userdata, file, ensure_ascii=False, indent=4)
            self.adduser.destroy()
        else:
            messagebox.showerror("Add Failed", "The username is incorrect, please try again.")
            self.text_box1.delete(0, tk.END)
            self.text_box3.delete(0, tk.END)
        
    def find(self):
        self.findpw = tk.Toplevel(root)
        self.findpw.geometry("280x150+610+320")
        self.findpw.title("Find...")
            
        label=tk.Label(self.findpw, text='Username:')
        label.place(x=30,y=15)
        self.text_box4 = tk.Entry(self.findpw, width=20)  
        self.text_box4.place(x=100,y=16) 
            
        button1 = tk.Button(self.findpw, text="Search",command=lambda:self.get_find())
        button1.place(x=50,y=100)
        button2 = tk.Button(self.findpw, text="Quit",command=lambda:self.close_window(self.findpw))
        button2.place(x=170,y=100)
        
    def get_find(self):
        name=self.text_box4.get().strip()
        password='0'
        for student in userdata['students']:
            if student.get('username') == name:
                password=student.get('password')
                messagebox.showinfo(title="Result", message=f"Password for {name} is: {password}")
        for teacher in userdata['teachers'] :
            if teacher.get('username') == name:
                password=teacher.get('password')
                messagebox.showinfo(title="Result", message=f"Password for {name} is: {password}")
        if password=='0':
            messagebox.showerror("Add Failed", "The username is incorrect, please try again.")
            self.text_box4.delete(0, tk.END)
        
    def close_window(self,window):
        window.destroy()
    
```
